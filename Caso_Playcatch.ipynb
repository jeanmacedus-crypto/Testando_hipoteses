Playcatch é uma plataforma de streaming e recentemente realizou uma atualização significativa em sua plataforma, introduzindo novos recursos e uma interface reformulada. A equipe de desenvolvimento está interessada em avaliar se essa atualização resultou em uma melhoria significativa nas avaliações dos usuários em comparação com uma plataforma concorrente.

```python
# carregando bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_avaliacoes = pd.DataFrame({
    'playcatch': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4],
    'concorrente': [3, 4, 3, 3, 2, 4, 3, 2, 3, 3, 4, 2, 3]
})
df_avaliacoes
```

A empresa deseja saber se há uma diferença estatisticamente significativa nas avaliações dos usuários entre a plataforma Playcatch e a concorrente após a atualização.

Como já esperamos, a equipe de análise construiu etapas para seguir e ajudar na tomada de decisão:




1 - Formule uma hipótese para o primeiro caso da Playcatch.

2 - Aplique o teste de hipótese que mais se encaixa com a natureza dos dados apresentados.

3 - Pelos resultados do teste, a hipótese nula formulada é rejeitada ou não rejeitada? Explique o que justifica sua decisão.

1 - Formulando hipótese

H0: há uma diferença estatisticamente significativa nas avaliações.

H1: não há uma diferença estatisticamente significativa nas avaliações.

```python
#2 aplique o teste de hipótese
#Neste cenário, o teste de Mann-Whitney seria apropriado, pois os dados representam avaliações de usuários em duas amostras independentes.

from scipy.stats import mannwhitneyu

stat, p_valor = mannwhitneyu(df_avaliacoes['playcatch'], df_avaliacoes['concorrente'])

print(f'Estatística do teste Mann-Whitney: {stat}')
print(f'Valor-p: {p_valor}')
```

```python
#3 Pelos resultados do teste, a hipótese nula formulada é rejeitada ou não rejeitada? Explique o que justifica sua decisão.

nivel_significancia=0.05

if p_valor < nivel_significancia:
    conclusao = "Rejeitar a hipótese nula"
else:
    conclusao = "Não rejeita a hipótese nula"
conclusao
```

como foi rejeitado a hipótese nula, afirmamos que **há uma diferença estatisticamente significativa nas avaliações.**

#Segundo caso Playcatch
Nesse novo caso, a Playcatch introduziu um novo recurso interativo em sua plataforma, chamado WatchTogether, que permite aos usuários assistir a conteúdo simultaneamente e compartilhar experiências em tempo real.

Para o estudo, a empresa coletou avaliações de usuários (em uma escala de 1 a 5 estrelas) antes e depois da introdução do novo recurso.

```python
df_wt = pd.DataFrame({
    'Antes_WatchTogether': [3, 4, 3, 4, 2, 4, 3, 2, 3, 3, 4, 2, 3],
    'Depois_WatchTogether': [4, 5, 3, 4, 5, 4, 4, 3, 5, 4, 5, 3, 4]
})

df_wt
```

A equipe de desenvolvimento deseja avaliar se a introdução desse novo recurso resultou em uma melhoria significativa nas avaliações dos usuários em comparação com as avaliações antes do lançamento do recurso. Para a resolução desse problema, a equipe de dados separou os seguintes passos:

1 - Formule uma hipótese para o segundo caso da Playcatch.

2- Aplique o teste de hipótese que mais se encaixa com a natureza dos dados apresentados.

3 - Pelos resultados do teste, a hipótese nula formulada é rejeitada ou não rejeitada? Explique o que justifica sua decisão.

1 - Formulando hipótese

H0: novo recurso resultou em uma melhoria significativa nas avaliações dos usuários.

H1: novo recurso não resultou em uma melhoria significativa nas avaliações dos usuários.

```python
#2 Aplicar o teste de hipótese
#Neste cenário, o teste de Wilcoxon seria apropriado, pois estamos comparando as avaliações dos mesmos usuários antes e depois da introdução do novo recurso.

from scipy.stats import wilcoxon

stat, p_valor = wilcoxon(df_wt['Antes_WatchTogether'], df_wt['Depois_WatchTogether'])

print(f'Estatística do teste de Wilcoxon: {stat}')
print(f'Valor-p: {p_valor}')
```

```python
#3 Resultados

nivel_significancia=0.05

if p_valor < nivel_significancia:
    conclusao = "Rejeitar a hipótese nula"
else:
    conclusao = "Não rejeita a hipótese nula"
conclusao
```

 rejeitamos a hipótese nula. Então, há uma diferença significativa nas avaliações médias antes e depois da introdução do WatchTogether.
